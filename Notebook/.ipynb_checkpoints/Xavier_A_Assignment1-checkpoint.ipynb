# **Individual Assingment**: Default Credit Cards Clients

A bank in you region want to build a model to predict credit card defaulters more accurately in order minimize money loss. For this task we have a set of data on default payments and demographic data to help us do our task. Data is comprised in the following CSV files:


**TRAINING**

**`train_customers.csv`**
 - `ID`: ID of each client
 - `LIMIT_BAL`: Amount of given credit in NT dollars (includes individual and family/supplementary credit
 - `SEX`: Gender (1=male, 2=female)
 - `EDUCATION`: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
 - `MARRIAGE`: Marital status (1=married, 2=single, 3=others)
 - `AGE`: Age in years
 
**`train_series.csv`**
 - `ID`: ID of each client
 - `MONTH`: The month to wich data is refering
 - `PAY`: Repayment status in the corresponding month (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, … 8=payment delay for eight months, 9=payment delay for nine months and above)
 - `BILL_AMT`: Amount of bill statement in the corresponding month (NT dollar)
 - `PAY_AMT`: Amount of previous payment in the corresponding month (NT dollar)
 
**`train_target.csv`**
 - `DEFAULT_JULY`: Default payment in July (1=yes, 0=no)
 
 
**TEST** (For PART 3)
 - **`test_data.csv`**
 
**SUBMISSION** (For PART 3)
 - **`submission_features.csv`**
 
**BACKUP**
 - **`train_data.csv`**

```python
# Pivoting table
train_series = pd.pivot_table(data_train_series, values=['PAY', 'BILL_AMT', 'PAY_AMT'], index=['ID'],
                    columns=["MONTH"] )
```

```python
# Joining and replacing Multi Index columns into one Index Column.
train_series.columns = ['_'.join(col).strip() for col in train_series.columns.values]
```

```python
from lazypredict.Supervised import LazyClassifier # Importing Lazy Classifier function for prediction
from sklearn.model_selection import train_test_split  # Importing train_test_split function
from sklearn.preprocessing import RobustScaler # Importing RobustScaler for scaling our data and removing outliers

# Dropping target and ID variable
X = df_train.drop(['DEFAULT_JULY','ID'], axis=1)
# Setting the target
y= df_train.DEFAULT_JULY
# Robust Scaling
scaler = RobustScaler()
X = scaler.fit_transform(X)
# Splitting our data
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.75,random_state =123)
# Fitting the models and displaying models results
clf = LazyClassifier(verbose=0,ignore_warnings=True)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

Based on the above table we selected the Light GBM model because is the faster and one of the most stable in it's metrics compared to the other models.

```python
# Hyperparameter Tuning and Cross Validation
from sklearn.model_selection import GridSearchCV # Importing GridSearchCV technique for finding the optimal parameter values with CV.
import lightgbm as lgb # Importing LightGBM model
```

```python
# LightGBM Parameters Grid
lgb_params = {
        'bagging_fraction': (0.5, 0.8),
        'bagging_frequency': (5, 8),
        'feature_fraction': (0.5, 0.8),
        'max_depth': (10, 13),
        'min_data_in_leaf': (90, 120),
        'num_leaves': (1200, 1550)
}

# Initialize a GridSearchCV object using 5-fold CV-
lgb_cv = GridSearchCV(estimator=lgb.LGBMClassifier(), scoring='roc_auc', param_grid=lgb_params, cv = 5,verbose=1)

# Train on training data-
lgb_cv.fit(X, y,verbose=1)
```

```python
# Importing performance and error metrics
from sklearn.metrics import f1_score,plot_roc_curve,roc_auc_score,plot_confusion_matrix,precision_recall_curve,classification_report,PrecisionRecallDisplay,confusion_matrix
```

```python
# Printing Best ROC AUC Score & Parameters
print(lgb_cv.best_score_, lgb_cv.best_params_)
# Importing LightGBM Best Estimator
model1=lgb_cv.best_estimator_
# Printing ROC Curve
plot_roc_curve(model1, X_test, y_test)
# Predicting with test data
y_pred = model1.predict(X_test)
# Printing Results
print("--------------------------------------")
print(classification_report(y_test,y_pred))
print("======================================")
```

```python
# Plotting Confusion Matriz
plot_confusion_matrix(model1,X_test,y_test);
plt.grid(None)
```

```python
# Importing Submission Data
df_submission = pd.read_csv('../data/submission_features.csv')
```

```python
# Preparing Submission Data for model
X_submission = df_submission.drop(['ID'], axis=1)
X_submission= scaler.fit_transform(X_submission)
# Soft Prediction of Submission Data
y_submission = model1.predict_proba(X_submission)[:, 1]
df_submission['probability'] = y_submission
# Final Submission Data Dataset
df_submission_final = df_submission[['ID','probability']]  
```

```python
# Defining Costs by each type of resuls (True Negatives, False Positives, False Negatives, True Positives) defined in the Question
def default_cost_function(tn, fp, fn, tp):
    cost = 0
    cost += (tn * -1000) 
    cost += (fp * 1000) 
    cost += (fn * 5000) 
    cost += (tp * -5000) 
    return cost
```

```python
# Creating Data Frame with multiple cost, recall, precision and threshold results
df_cost = pd.DataFrame({'precision':precisions[:-1], 'recall':recalls[:-1], 'threshold':thresholds, 'cost':costs, 'tn':tns, 'fp':fps, 'fn':fns, 'tp':tps})
```

```python
# Getting the min cost value
min_cost = df_cost['cost'].min()
```
